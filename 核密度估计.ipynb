# 理解 撰写高斯核密度函数 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 设置字体
plt.rcParams['font.family'] = ['SimHei']

# 交易金额数据
data = np.array([10, 15, 20, 22, 25, 28, 30, 35, 40, 50, 55, 60, 200])

# 计算核密度估计
kde = gaussian_kde(data)
x_range = np.linspace(0, 250, 1000)  # 生成 x 轴的范围
density = kde(x_range)  # 计算每个 x 值的密度

# 绘制 KDE 曲线
plt.figure(figsize=(8, 4))
plt.plot(x_range, density, label="KDE 密度估计")
plt.axvline(200, color='red', linestyle='--', label="疑似异常点")
plt.xlabel("交易金额 ($)")
plt.ylabel("概率密度")
plt.title("信用卡交易的 KDE 密度估计")
plt.legend()
plt.show()
```

```python
print("Dataset used for KDE:", kde.dataset)
print("Bandwidth factor:", kde.factor)
```
